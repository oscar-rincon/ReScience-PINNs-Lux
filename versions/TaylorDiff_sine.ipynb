```julia
using TaylorDiff, Plots 

# Define the range and function with Float32
x = collect(range(0f0, 5f0, length=200))
x = reshape(x, 1, :) 

# Define the function
f(x) = sin.(x) 

# Calculate the first derivative
derivative = TaylorDiff.derivative.(f, x, Float32.(ones(size(x))), Val(2))

# Plot the function and its derivative
plot(x', derivative')
plot!(x', f.(x)')
 
```
